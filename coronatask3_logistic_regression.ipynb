```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
c = pd.read_csv('/content/covid-19 symptoms dataset (1).csv')
```

```python
c.head()
```

```python
c.isnull().sum()
```

```python
c.info()
```

```python
ind = c[['fever','bodyPain','runnyNose','diffBreath']]

dep = c['infectionProb']
```

```python
LR = LogisticRegression()
```

```python
LR.fit(ind,dep)
```

```python
fever = int(input('Enter fever temperature:'))
body_pain = int(input('have body pain or not?'))
run_nose = int(input('have runny nose?'))
diff_breadth = int(input('have difficulty in breadth?'))

print(LR.predict([[fever,body_pain,run_nose,diff_breadth]]))
```

```python
from sklearn.metrics import accuracy_score
pv = LR.predict(ind)
accuracy_score(dep,pv)
```

```python
LR.score(ind,dep)
```
